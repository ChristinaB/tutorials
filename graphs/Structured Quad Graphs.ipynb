```python
%matplotlib inline
import numpy as np

from landlab.graph import DualRectilinearGraph
from landlab.plot.graph import plot_graph
```

```python
y = np.linspace(0, 50, 5)
x = np.logspace(1, 2, 4)
graph = DualRectilinearGraph((y, x))
```

```python
plot_graph(graph, at='node,link,patch')
plot_graph(graph, at='corner,face,cell')
```

```python
%matplotlib inline
import numpy as np

from landlab.graph import DualStructuredQuadGraph
from landlab.plot.graph import plot_graph
```

```python
rho, theta = np.meshgrid(np.linspace(1, 5, 3), np.linspace(0, np.pi * .75, 5))
y, x = rho * np.sin(theta), rho * np.cos(theta)

graph = DualStructuredQuadGraph((y, x), shape=y.shape)
```

```python
plot_graph(graph, at='node,link,patch')
plot_graph(graph, at='corner,face,cell')
```
